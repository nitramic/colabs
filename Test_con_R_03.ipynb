Comienzo con la carga de datos desde yahoo finance:

```r
mercado = "YPF"
StartDay ="2022-10-01"
StopDay = "2023-08-25"
```

```r
#Cargo los datos en la funcion
valor = getSymbols(mercado,auto.assign = FALSE,from = StartDay, to = StopDay)
```

Genero el primer grafico con candleChart

```r
candleChart(valor)
```

```r
return = dailyReturn(valor, type = "arithmetic")
charts.PerformanceSummary(return)
```

```r
table.Drawdowns(return)
```

```r
table.CalendarReturns(dailyReturn(to.monthly(valor),type = "arithmetic"))
```
